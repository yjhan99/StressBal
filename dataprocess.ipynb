```python
import pandas as pd
import os
```

```python
os.system('mkdir -p data/train/0')
os.system('mkdir -p data/train/1')
```

```python
df = pd.read_csv('traindata.csv')
```

```python
df.columns
```

```python
partition_train_f = open('data/partition_train.txt', 'w')

for index, row in df.iterrows():
    if row['label'] == 1:
        f = open('data/train/1/sample_'+str(index)+'.txt', 'w')
        partition_train_f.write('data/train/1/sample_'+str(index)+'.txt\n')
    elif row['label'] == 0:
        f = open('data/train/0/sample_'+str(index)+'.txt', 'w')
        partition_train_f.write('data/train/0/sample_'+str(index)+'.txt\n')
    data = ''
    for column in df.columns:
        data += str(row[column])
        data += ','
    f.write(data[:-1])
    f.close()

partition_train_f.close()
```
